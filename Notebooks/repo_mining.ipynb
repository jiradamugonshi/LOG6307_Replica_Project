```python
import pandas as pd
column_names = ['dir', 'commitId', 'author', 'commitDate']
csvFile = pd.read_csv('output_openstack.csv', delimiter=";", header=None, names=column_names)
csvFile['commitDate'] = pd.to_datetime(csvFile['commitDate'])
```

```python
csvFile['month_year'] = csvFile['commitDate'].dt.to_period('M')
```

```python
unique_values = csvFile['dir'].unique()
unique_values
```

```python
csvFile.head(10)
```

```python
grouped = csvFile.groupby(['dir', 'month_year']).size().reset_index(name='commits_per_month')
```

```python
grouped.head(30)
```

```python
threshold = 2

# Group by dir and filter out groups where any value in commits_per_month is below the threshold
to_remove = grouped.groupby('dir').filter(lambda x: (x['commits_per_month'] < threshold).any())

# Keep only the groups that don't meet the condition
filtered_grouped = grouped[~grouped['dir'].isin(to_remove['dir'].unique())]
```

```python
filtered_grouped.head(10)
```

```python
unique_values = filtered_grouped['dir'].unique()

print(unique_values)
```
