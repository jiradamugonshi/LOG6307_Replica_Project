```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
```

```python
csvFile = pd.read_csv('IST_WIK.csv')
```

```python
csvFile.head(10)
```

```python
X = csvFile.iloc[:, 2:14].values
y = csvFile.iloc[:, 14].values
```

```python
# define the model
model = RandomForestClassifier()
# fit the model
model.fit(X, y)
# get importance
importance = model.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
 print('Feature: %0d, Score: %.5f' % (i,v))
```

```python
clf = RandomForestClassifier()
importances = []
column_names_list = csvFile.columns[2:-1].tolist()

for _ in range(10):    
    clf.fit(X, y)    
    importances.append(clf.feature_importances_)


importances_df = pd.DataFrame(importances, columns=column_names_list)

# Calculate the median of feature importances
median_importances = importances_df.median().sort_values(ascending=False)

# Display the median feature importance for each feature
print(median_importances)
    
```

```python
importances_df.to_csv('feature_importance.csv', columns=column_names_list, index=False)
```
